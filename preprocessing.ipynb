```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the dataset (replace 'your_dataset.csv' with your file path)
df = pd.read_csv("E:\Ai project\smart_mart_inventory.csv")

# 1. Display initial information
print("Initial Data:")
print(df.head())
print("\nData Types:")
print(df.dtypes)

# 2. Handle Missing Values
# Replace the string "N/A" in 'PrescriptionRequired' with a more appropriate value (e.g., False)
df['PrescriptionRequired'] = df['PrescriptionRequired'].replace("N/A", None)
# Alternatively, if you want to assume N/A means not required, you might fill with False
df['PrescriptionRequired'] = df['PrescriptionRequired'].fillna("Not Required")

# 3. Remove Duplicates
df.drop_duplicates(inplace=True)

# 4. Convert 'ExpirationDate' to datetime
df['ExpirationDate'] = pd.to_datetime(df['ExpirationDate'])

# 5. Ensure Numeric Data Types for relevant columns
df['Price'] = df['Price'].astype(int)
df['Quantity'] = df['Quantity'].astype(int)
df['UnitsSoldLastMonth'] = df['UnitsSoldLastMonth'].astype(int)
df['StockThreshold'] = df['StockThreshold'].astype(int)

# 6. Scale / Normalize Numerical Columns
scaler = MinMaxScaler()
df[['Price', 'Quantity', 'UnitsSoldLastMonth']] = scaler.fit_transform(df[['Price', 'Quantity', 'UnitsSoldLastMonth']])

# 7. Encoding Categorical Variables
# OneHotEncode MainCategory and SubCategory
df = pd.get_dummies(df, columns=['MainCategory', 'SubCategory'])

# Optional: If you want to encode 'PrescriptionRequired' if needed (though it may be text)
# For demonstration, let's leave it as it is, or you could map it:
# df['PrescriptionRequired'] = df['PrescriptionRequired'].map({"Not Required": 0, "Required": 1})

# 8. Display the cleaned and preprocessed DataFrame
print("\nPreprocessed Data:")
print(df.head())
```
